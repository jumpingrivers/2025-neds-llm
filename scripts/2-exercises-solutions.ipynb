## Question 1

```python
chat.chat("Who were the key players in 1066?")
```

## Question 1a

Use a structured query to find out the name of the victor in 1066

```python
from pydantic import BaseModel, Field


class Person(BaseModel):
    name: str = Field(description="The person's name")


chat.chat_structured(
  "Who was the winner of 1066?",
  data_model = Person
)
```

## Question 1b

Use a structured query to find out the name, age and nationality of the victor of 1066

```python
class Person(BaseModel):
    name: str = Field(description="The person's name")
    age: int = Field(description="The person's age in years")
    nationality: str = Field(description="The person's nationality")


chat.chat_structured(
  "Who was the winner of 1066?",
  data_model = Person
)
```

## Question 1c (harder)

Generate a list of the contenders for the English throne in 1066

_Hint_: Add your data class from 1b and rename as Person. It will then be wrapped in the People class defined below

```python
class Person(BaseModel):
    name: str = Field(description="The person's name")
    age: int = Field(description="The person's age in years")
    nationality: str = Field(description="The person's nationality")


class People(BaseModel):
    persons: list[Person]


chat.chat_structured(
  "Who were the key players in 1066?",
  data_model = People
)
```

# Question 2

## Question 2a

Get a summary of what is shown in the image

## Question 2b

Now use a structured query to get the following info for each species shown:

- Name of the species

- Average weight of the species in kilograms

- Typical diet for the species

- Average lifespan of the species in years

```python
class Animal(BaseModel):
    name: str = Field(description="Name of the species")
    weight: float = Field(description="Average weight of the species in kilograms")
    diet: str = Field(description="Typical diet of the species")
    lifespan: str = Field(description="Average lifespan of the species in years")


class Animals(BaseModel):
    animals: list[Animal]


chat.chat_structured(
  "What are the unique animal species in this image?",
  chatlas.content_image_file(image_filepath),
  data_model = Animals
)
```
