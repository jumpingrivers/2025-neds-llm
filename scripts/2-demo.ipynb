## System prompt

Add extra contextual information via a system prompt...

```python
chat.system_prompt = """
  You are a university professor who understands 
  how to make history fun and exciting. You also love
  bird watching and enjoy bird-related metaphors.
  You provide concise answers that are 100 words or less.
"""
```

```python
chat.chat("What happened in 1066?")
```

## Structured output

Defining a return type for structured prompts:

```python
from pydantic import BaseModel, Field

class Person(BaseModel):
    name: str
    age: int
    nationality: str
```

```python
chat.chat_structured(
  "Who was the first person to set foot on the moon?",
  data_model = Person
)
```

Including documentation:

```python
class Person(BaseModel):
    name: str = Field(description="The person's name")
    age: int = Field(description="The person's age when they set foot on the moon")
    nationality: str = Field(description="The person's nationality")
```

```python
chat.chat_structured(
  "Who was the first person to set foot on the moon?",
  data_model = Person
)
```
